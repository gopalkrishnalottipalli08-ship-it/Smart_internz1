```python
import os
import shutil
import pandas as pd

# Paths
train_path = "train"
labels_path = "labels.csv"
output_path = "train_data"

# Read labels
labels = pd.read_csv(labels_path)

# Create main output folder
os.makedirs(output_path, exist_ok=True)

# Create breed folders
for breed in labels['breed'].unique():
    os.makedirs(os.path.join(output_path, breed), exist_ok=True)

# Copy images into respective breed folders
for index, row in labels.iterrows():
    image_id = row['id']
    breed = row['breed']
    
    source = os.path.join(train_path, image_id + ".jpg")
    destination = os.path.join(output_path, breed, image_id + ".jpg")
    
    if os.path.exists(source):
        shutil.copyfile(source, destination)

print("Images organized successfully!")
```

```python
total_images = 0

for folder in os.listdir(output_path):
    folder_path = os.path.join(output_path, folder)
    total_images += len(os.listdir(folder_path))

print("Total images in class folders:", total_images)
print("Total class folders:", len(os.listdir(output_path)))
```

```python
from tensorflow.keras.applications.vgg19 import VGG19
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Model

Image_size = [224, 224]

base_model = VGG19(
    input_shape=Image_size + [3],
    weights='imagenet',
    include_top=False
)

for layer in base_model.layers:
    layer.trainable = False

y = Flatten()(base_model.output)

# IMPORTANT CHANGE HERE
final = Dense(generator.num_classes, activation='softmax')(y)

vgg19_model = Model(inputs=base_model.input, outputs=final)

vgg19_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
vgg19_model.fit(generator, epochs=6)
```
